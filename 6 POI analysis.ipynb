# 6: Point of interest in the area Analysis 

###### Gets Point Of Interests in given radius + location and generates relevant maps for analysis. Input of location can be either coordinates or node ID (as can be determined from the dataframes in previous scripts).

### Setting up packages and variables

```
import osmnx as ox
import matplotlib.pyplot as plt
%matplotlib inline
from pyproj import CRS

# theme colors
miev_pink = '#f06675'
miev_blue = '#2c3174'
miev_yellow = '#fbe56d'

# # from node ID
# place = 'Binnenstad Delft, Netherlands'
# G = ox.graph_from_place(place, buffer_dist=1000)
# node_id = 1436427125 
# G.nodes[node_id]
# x = G.nodes[node_id]['x'] #lon
# y = G.nodes[node_id]['y'] #lat

#from coordinates
#hub1
x = 4.354905104373707
y = 52.00839587659508

# #hub2
# x = 4.3705270202481055
# y = 52.01112279261438

# #hub3
# x = 4.361283532900554
# y = 52.017343387857906
```

### Download data from enviroment

```
# Get data 500m radius
G = ox.geometries_from_point((y,x), dist=500, tags={'highway':True, 'landuse':True, 'building':True, 'amenity':True, 'leisure':True, 'tourism': True})
```

### Landuse Graph

```
# Graph plot
G.to_crs(epsg=27700, inplace=True)

fig,ax = plt.subplots(figsize=(16,16))
G[G['landuse'].notna()].plot(ax=ax, column='landuse', cmap='spring', legend=True)
G[G['highway'].notna()].plot(ax=ax, color='white')
# G[G['building'].notna()].plot(ax=ax, color='dimgrey', zorder=3)

ax.set_facecolor(miev_blue)
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)

plt.savefig('data/landuse.png')
```

### Buildings Graph

```
# Graph plot
G.to_crs(epsg=27700, inplace=True)

fig,ax = plt.subplots(figsize=(16,16))
G[G['highway'].notna()].plot(ax=ax, color='white')
G[G['building'].notna()].plot(ax=ax, column='building', cmap='tab20', legend=True, zorder=3)

ax.set_facecolor(miev_blue)
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)

plt.savefig('data/buildings.png')
```

### Amenities Graph

```
# Graph plot
G.to_crs(epsg=27700, inplace=True)

fig,ax = plt.subplots(figsize=(16,16))
G[G['highway'].notna()].plot(ax=ax, color='white')
G[G['amenity'].notna()].plot(ax=ax, column='amenity', cmap='tab20', legend=True, zorder=3)

ax.set_facecolor(miev_blue)
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)

plt.savefig('data/amenities.png')
```
