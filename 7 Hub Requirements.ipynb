# 6: Get HUB requirements for given location

###### Gets amount of type of houses in a given place and area and calculates amount of cars based on the input norm per house and apartment. Includes an estimation of amount of shared cars that this amount can be replaced with to determine the HUB size. Input of location can be either coordinates or node ID.

### Setting up packages and variables

```
import osmnx as ox
import math

# Edit based on location
# Ratio car to houses (source: Nota Parkeernormen 2018 Delft)
apartment_to_car = 0.5
house_to_car = 1.3
# Ratio car to shared car (source: Transport & Enviroment 2017)
car_to_sharedcar = 5

# theme colors
miev_pink = '#f06675'
miev_blue = '#2c3174'
miev_yellow = '#fbe56d'

# # from node ID
# place = 'Binnenstad Delft, Netherlands'
# G = ox.graph_from_place(place, buffer_dist=1000)
# node_id = 1436427125 
# G.nodes[node_id]
# x = G.nodes[node_id]['x'] #lon
# y = G.nodes[node_id]['y'] #lat

#from coordinates
#hub1
x = 4.354905104373707
y = 52.00839587659508

# #hub2
# x = 4.3705270202481055
# y = 52.01112279261438

# #hub3
# x = 4.361283532900554
# y = 52.017343387857906
```

```
# Get data 500m radius
apartments = ox.geometries_from_point((y,x), dist=500, tags={"building": "apartments"})
house = ox.geometries_from_point((y,x), dist=500, tags={"building": "house"})
residential = ox.geometries_from_point((y,x), dist=500, tags={"building": "residential"})
dormitory = ox.geometries_from_point((y,x), dist=500, tags={"building": "dormitory"})
```

```
# print amounts 
print("There are", len(apartments), "apartments,", len(house),"houses,", len(residential), "residential buildings and", len(dormitory),"dormitories nearby.")

# multiply by household to car ratio and divide by shared car ratio
total_apartment = len(apartments) + len(residential) + len(dormitory)
total_houses = len(house)
total_cars_house = total_houses * house_to_car
total_cars_apartments = total_apartment * apartment_to_car
total_cars = total_cars_house + total_cars_apartments
total_sharedcars = total_cars / car_to_sharedcar
total_sharedcars = math.ceil(total_sharedcars)

print("There will be an estimated need for a maximum of", total_sharedcars, "shared cars in this area.")
```
